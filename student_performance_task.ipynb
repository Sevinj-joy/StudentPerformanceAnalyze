Korelasiya üçün uyğunlaşdırma:


```python
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})
df = pd.get_dummies(df, columns=['Distance_from_Home'], drop_first=False)
df['Internet_Access'] = df['Internet_Access'].map({'No': 0, 'Yes': 1})
```

```python
import scipy.stats as stats
# Korelasiya hesablamaq
# Spearman Korrelyasiyası
spearman_gender_distance = stats.spearmanr(df['Gender'], df['Distance_from_Home_Far'])
spearman_internet_gender = stats.spearmanr(df['Gender'], df['Internet_Access'])
spearman_distance_internet = stats.spearmanr(df['Distance_from_Home_Far'], df['Internet_Access'])

# Nəticələri çap edirik
print("Spearman Korrelyasiyası (Gender ilə Distance_from_Home_Far):", spearman_gender_distance.correlation)
print("Spearman Korrelyasiyası (Gender ilə Internet_Access):", spearman_internet_gender.correlation)
print("Spearman Korrelyasiyası (Distance_from_Home_Far ilə Internet_Access):", spearman_distance_internet.correlation)
import matplotlib.pyplot as plt

# Define correlation values
correlations = [
    spearman_gender_distance.correlation,
    spearman_internet_gender.correlation,
    spearman_distance_internet.correlation
]

# Define labels for each pair of variables
labels = [
    'Gender & Distance_from_Home_Far',
    'Gender & Internet_Access',
    'Distance_from_Home_Far & Internet_Access'
]

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(labels, correlations, color=['skyblue', 'salmon', 'lightgreen'])

# Add labels and title
plt.xlabel("Variable Pairs")
plt.ylabel("Spearman Correlation Coefficient")
plt.title("Spearman Correlation between Variable Pairs")
plt.ylim(-1, 1)  # Spearman correlation range is from -1 to 1

# Show the values on top of each bar
for i, v in enumerate(correlations):
    plt.text(i, v + 0.05 if v >= 0 else v - 0.05, f"{v:.2f}", ha='center')

# Display the plot
plt.show()
```

   - Korelasiya dəyəri 0-a çox yaxın olduğu üçün **Gender** və **Distance_from_Home_Far** arasında heç bir əhəmiyyətli monotonic əlaqə yoxdur. Yəni, kişinin və ya qadının evinin uzaqlığı ilə əlaqədar xüsusi bir nümunə müşahidə edilmir.

### 2. **Spearman Korrelyasiyası (Gender ilə Internet_Access):**

   - Bu nəticə də çox zəif bir əlaqəni göstərir. Korelasiya dəyəri **0**-a çox yaxın olduğu üçün **Gender** ilə **Internet_Access** (İnternetə giriş) arasında da heç bir əhəmiyyətli əlaqə yoxdur.

### 3. **Spearman Korrelyasiyası (Distance_from_Home_Far ilə Internet_Access):**

   - Həm **evin uzaqlığı**, həm də **internetə giriş** arasında bu qədər kiçik və zəif bir korelasiya olduğu halda, bu əlaqə statistik olaraq əhəmiyyətsizdir.

```python
df_sample = data.sample(n=500, random_state=42)
```

```python
exam_scores = df_sample['Exam_Score']
```

```python
#  (mean) və (standard error) 
mean_exam_score = exam_scores.mean()
std_error = exam_scores.std() / np.sqrt(len(exam_scores))
```

```python
confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha / 2, df=len(exam_scores) - 1)
```

```python

margin_of_error = t_critical * std_error
lower_bound = mean_exam_score - margin_of_error
upper_bound = mean_exam_score + margin_of_error

print(f"Orta dəyər: {mean_exam_score:.2f}")
print(f"Etibarlılıq intervalı: [{lower_bound:.2f}, {upper_bound:.2f}]")
```

İzah:

Orta dəyərin (mean) və standart səhvin (standard error) hesablanması:

Orta dəyər: mean_exam_score = exam_scores.mean()
Standart səhv: std_error = exam_scores.std() / np.sqrt(len(exam_scores))
exam_scores.std() nümunə üzrə standart sapmanı, np.sqrt(len(exam_scores)) isə nümunə ölçüsünün kvadrat kökünü göstərir.
t-distribusiyasından istifadə edərək 95%-lik etibarlılıq intervalının hesablanması:

t.ppf(1 - alpha / 2, df=len(exam_scores) - 1) t-distribusiyasının kritik dəyərini (t- критический) tapır.
alpha 0.05-ə bərabərdir, çünki 95%-lik etibarlılıq intervalı üçün alpha = 1 - 0.95 = 0.05.
